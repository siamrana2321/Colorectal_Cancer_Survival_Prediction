```python
import numpy as np # linear alg
import pandas as pd
```

```python
df = pd.read_csv("./dataset/colorectal_cancer_prediction.csv")
```

```python
print(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Display basic info about the dataset
print("Dataset Info:")
df.info()

# Display summary statistics
print("\nSummary Statistics:")
print(df.describe())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Display the first few rows of the dataset
print("\nFirst Few Rows:")
print(df.head())

# Distribution of Age
df['Age'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Countplot for Gender
sns.countplot(x='Gender', data=df, palette='coolwarm')
plt.title("Gender Distribution")
plt.show()

# Countplot for Stage at Diagnosis
sns.countplot(x='Stage_at_Diagnosis', data=df, palette='viridis')
plt.title("Stage at Diagnosis Distribution")
plt.show()

# Countplot for Survival Status
sns.countplot(x='Survival_Status', data=df, palette='Set2')
plt.title("Survival Status Distribution")
plt.show()

# Correlation heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()

# Boxplot for Age vs. Survival Status
plt.figure(figsize=(8, 4))
sns.boxplot(x='Survival_Status', y='Age', data=df, palette='muted')
plt.title("Age vs. Survival Status")
plt.show()

# Barplot for Treatment Access
sns.countplot(x='Treatment_Access', hue='Survival_Status', data=df, palette='pastel')
plt.title("Treatment Access vs. Survival Status")
plt.show()

# Boxplot for Time to Recurrence vs. Stage at Diagnosis
plt.figure(figsize=(8, 4))
sns.boxplot(x='Stage_at_Diagnosis', y='Time_to_Recurrence', data=df, palette='coolwarm')
plt.title("Time to Recurrence vs. Stage at Diagnosis")
plt.show()
```
